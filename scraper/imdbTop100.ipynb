```python
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome('/Applications/chromedriver')
```

```python
driver.get('https://www.imdb.com/list/ls055592025/')
```

```python
movies = driver.find_elements(By.XPATH,'//div[@class = "lister-item mode-detail"]')
```

```python
len(movies)
```

```python
df = pd.DataFrame(columns=['Title','Synopses', 'Genre'])
```

```python
df
```

```python
for i in range(1,len(movies)+1):
    movie_info = {'Title':'','Synopses':'','Genre':''}
    movie_info['Title'] = driver.find_element(By.XPATH, '//div[@class = "lister-item mode-detail"]['+str(i)+']//h3[@class="lister-item-header"]//a').text
    movie_info['Synopses'] = driver.find_element(By.XPATH, '//div[@class = "lister-item mode-detail"]['+str(i)+']//div[@class="lister-item-content"]//p[@class=""]').text
    movie_info['Genre'] = driver.find_element(By.XPATH, '//div[@class = "lister-item mode-detail"]['+str(i)+']//div[@class="lister-item-content"]//span[@class="genre"]').text
    df = df.append(movie_info, ignore_index=True)
```

```python
df
```

```python
df.to_csv('output/top100desc.csv')
```
